## Need for stemming and Lemmatization
Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem. Lemmatization groups together different inflected forms of a word, called lemma. Somehow similar to stemming as it maps several words into one common root. Output of lemmatisation is a proper word.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

## About Wordnet
WordNet is a lexical database of English. Using synsets, helps find conceptual relationships between words such as hypernyms, hyponyms, synonyms, antonyms etc.

<i>Reference:https://www.nltk.org/api/nltk.corpus.reader.html#nltk.corpus.reader.wordnet.Lemma</i>

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

Tagging is a kind of classification that may be defined as the automatic assignment of description to the tokens.
Here the descriptor is called tag,which may represent one of the part-of-speech, semantic information and so on.

```python
from nltk import pos_tag
w = "better"
pos_tag([w])
```

```python
##We must exclude the stop words before training the data 
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops, string.punctuation
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
training_documents_orig = documents
testing_documents_orig= test_documents

training_documents = documents[0:7000]
testing_documents = documents[7000:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
import nltk
freq = nltk.FreqDist(all_words_orig)
common = freq.most_common(3000)
features_orig = [i[0] for i in common]
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```
